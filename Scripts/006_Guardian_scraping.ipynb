```python
def get_urls(url:str, year, month, day) -> list:
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')

    urls = pd.DataFrame(columns = ['urls'])
    for link in soup.find_all('a'):
        urls.loc[len(urls)] = [link.get('href')]
        
    urls = urls[urls['urls'].str.contains('https://www.theguardian.com/uk-news/{}/{}/{}/'.format(year,month,"%02d" % day), na=False)]
    urls = urls.drop_duplicates()
    return urls
```

```python
def get_article_text(url:str) -> list:
    """Parse guardian article and return text in list of strings"""
    
    article = requests.get(url)
    soup = BeautifulSoup(article.content, "html.parser")
    body = soup.find(id='maincontent')
    text = [p.text for p in body.find_all("p")] 
    full_text = ' '.join(text)
    return full_text
```
